This is a short examination of some data concerning sea levels over the period 1993-2021, data is from Kaggle https://www.kaggle.com/kkhandekar/global-sea-level-1993-2021

The overall conclusion of the analysis is that global mean sea levels have risen over the period, rising by just under 100mm. All measurements are in mm, more information can be found on the Kaggle page of the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sealevel.csv")
```

```python
df.head()
```

GMSL is Global Mean Sea Level, GIA is Global Isostatic Adjustment (this is the process of land springing back up after being previously covered by glaciers)

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
num_data = ['TotalWeightedObservations', 'GMSL_noGIA', 'StdDevGMSL_noGIA',
       'SmoothedGSML_noGIA', 'GMSL_GIA', 'StdDevGMSL_GIA', 'SmoothedGSML_GIA',
       'SmoothedGSML_GIA_sigremoved']
for i, col in enumerate(num_data):
    plt.figure(i)
    sns.distplot(df[col], bins =1000)
```

```python
for i, col in enumerate(['GMSL_noGIA','GMSL_GIA']):
    plt.figure(i)
    sns.boxplot(df[col])
```
